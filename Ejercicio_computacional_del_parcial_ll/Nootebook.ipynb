```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, uniform, gamma, cauchy
```

$En\ este \ ejercicio \ se \ demostrara \ el \ teorema \ del \ límite \ central$

```python
# Array con el tipo de muestreos y numero de muestras
N = np.array([[1e4, 10], 
              [1e5, 50], 
              [1e6, 100]])
```

# Exponencial

```python

fig, ax = plt.subplots(figsize=(8,6))

#Tenemos  que lambda es 1/miu entonces los valores de miu para los lambda dados son: 100
fig.suptitle('Exponencial con $\\lambda = 0.01$', fontsize=16)

for i in range(3): # recorremos sobre una cantidad de muestras
    n = int(N[i,0]) # tipo de muestreo
    m = int(N[i,1]) # numero de muestras de cada tipo

    muestras = expon.rvs(scale=100, size=(n, m)) #Calculamos la muestra

    # Promedios por fila
    muestras_prom = muestras.mean(axis=1)

    # Histograma (frecuencia normalizada)
    ax.hist(
        muestras_prom,
        bins=50,
        range=(0,500),
        density = True,
        alpha=0.5,
        edgecolor='black',
        label=f"n={n}, m={m}"
    )

# Curva gaussiana (como exige el taller)
x = np.linspace(0, 500, 500)


y = norm(loc=100, scale=100/np.sqrt(m)).pdf(x) 
ax.plot(x, y, 'r-', lw=2, label=f"Gaussiana ($\\mu$=100, $\\sigma$=100)")

ax.set_xlabel("Valores")
ax.set_ylabel("Frecuencia normalizada")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.savefig("Imagenes/exponencial_taller.png")
plt.show()
```

# Cauchy

```python

fig, ax = plt.subplots(figsize=(8,6))

fig.suptitle('Lorentziana con $x_0 = 100$, $\\gamma = 100$', fontsize=16)

for i in range(3): # recorremos sobre una cantidad de muestras
    n = int(N[i,0]) # tipo de muestreo
    m = int(N[i,1]) # numero de muestras de cada tipo

    muestras = cauchy.rvs(loc=100, scale=100, size=(n, m)) #Calculamos la muestra

    # Promedios por fila
    muestras_prom = muestras.mean(axis=1)

    # Histograma (frecuencia normalizada)
    ax.hist(
        muestras_prom,
        bins=50,
        range=(0,500),
        density = True,
        alpha=0.5,
        edgecolor='black',
        label=f"n={n}, m={m}"
    )

# Curva gaussiana (como pide el taller)
x = np.linspace(0, 500, 500)


y = norm(loc=100, scale=100/np.sqrt(m)).pdf(x) 
ax.plot(x, y, 'r-', lw=2, label=f"Gaussiana ($\\mu$=100, $\\sigma$=100)")

ax.set_xlabel("Valores")
ax.set_ylabel("Frecuencia normalizada")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.savefig("Imagenes/lorentziana_taller.png")
plt.show()
```
