```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2,cauchy, norm
```

# **Chi-cuadrado**

```python
#se crea un array de n
ns=np.array([1,10,100,1000]) # número de grados de libertad

#se define un diccionario para guardar los datos
resultados={}
```

```python


m = int(1e5)  # número de muestras
plt.figure(figsize=(7,4))
plt.title('Distribución de Chi cuadrado para distintos GL')
for n in ns:
    # Generar muestras
    samples = chi2.rvs(df=n, size=m)

    # Graficar histograma
    plt.hist(samples, bins=np.linspace(0,2000, 200), density=True, alpha=0.6, label =f"Chi-cuadrado con {n} GL")

plt.ylim(0,0.1)
plt.xlabel('valores')
plt.ylabel('Frecuencia normalizada')
plt.legend()    
plt.grid()
plt.show()
```

```python


# Generaciones en [0, 2000]
m = int(1e5)
x = np.linspace(0, 2000, m)

ns = [1, 10, 100, 1000]

plt.figure(figsize=(7,4))
plt.title('Distribución Chi-cuadrado para distintos GL')

for n in ns:
    y = chi2.pdf(x, df=n)
    plt.plot(x, y, label=f"χ² (n={n})")

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

## Función de Lorentz

```python


m = int(1e5)  # número de muestras

# Parámetros (x0, gamma)
params = [(1,2), (10,20), (100,200), (1000,200)]

plt.figure(figsize=(8,5))
plt.title('Distribución de Lorentz (Cauchy) para distintos parámetros')

for x0, gamma in params:
    # Generar muestras
    samples = cauchy.rvs(loc=x0, scale=gamma, size=m)
    
    # Graficar histograma
    plt.hist(samples, bins=np.linspace(0,2000, 200), range=(x0 - 5*gamma, x0 + 5*gamma), 
             density=True,
             label = f'X0 = {x0}  gamma = {gamma}')

# Ajustar visualización (muy importante por colas largas)
plt.ylim(0, 0.04)
plt.xlabel('valores')
plt.ylabel('Frecuencia normalizada')
plt.legend()
plt.grid()

plt.show()
```

## Distribución normal

```python


m = int(1e5)  # número de muestras

# Parámetros (mu, sigma^2)
params = [(1,2), (10,20), (100,200), (1000,200)]

plt.figure(figsize=(8,5))
plt.title('Distribución Normal para distintos parámetros')

for mu, var in params:
    sigma = np.sqrt(var)  # convertir varianza → desviación estándar
    
    # Generar muestras
    samples = norm.rvs(loc=mu, scale=sigma, size=m)
    
    # Definir rango razonable (±4σ cubre casi todo)
    xmin = mu - 4*sigma
    xmax = mu + 4*sigma
    
    # Histograma
    plt.hist(samples, bins = np.linspace(0,2000, 100), range=(xmin, xmax),
             density=True, alpha=0.5,
             label=f"(μ={mu}, σ²={var})")
plt.ylim(0, 0.15)
plt.xlabel('valores')
plt.ylabel('Frecuencia normalizada')
plt.legend()
plt.grid()

plt.show()
```
