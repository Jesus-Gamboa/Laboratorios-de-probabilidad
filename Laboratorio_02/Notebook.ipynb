```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2,cauchy, norm
```

# **Chi-cuadrado**

```python
#se crea un array de n
n=np.array([1,10,100,1000])
#tamaño de la muestra
muestra=100000
#se define un diccionario para guardar los datos
resultados={}
```

```python
#ciclo for para encontrar la densidad de probabilidad
for i in n:
  #datos con el grado de libertad i
  resultados[f"n={n}"]=chi2.rvs(df=i,size=muestra)
```

```python
#gráfica
#tamaño
plt.figure((10,6))
#ciclo for para realizar la gráfica
for i in n:
  plt.hist(muestras[f"n={n}"], bins=200, density=True, alpha=0.5, label=f"n={n}")

plt.title("Distribución Chi-cuadrado para distintos n")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```
